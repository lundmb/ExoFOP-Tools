# ExoFOP-TESS php scripts

A number of php scripts are available for more easily accessing data. Broadly, these can be separated into those that that will download files and those that will return text. Summaries of scripts are presented below, and the path for all scripts begins with "https://exofop.ipac.caltech.edu/tess/"

Some php scripts have additional terms that may be used. For a php call with additional parameters, a question mark follows '.php' and is followed by the first parameter. In the case of multiple parameters, these should be separated by '&'.

For those that return text, there are a few formatting options that may be available. They are described here, and each entry will note which of these options is available.

 - text : Returns columns separated by white space
 
 - csv : Returns csv formatted text
 
 - pipe : Returns pipe-delimited text
 
 
Some of these php scripts already have been incorporated into Python functions that have been developed to make these more quickly accessible or to perform more complicated operations. These will be explained in additional notebooks and allow for quicker integration into Python.
 

### Accessing proprietary period data

The ExoFOP enables users to upload data with a 12-month proprietary period as part of the TESS Follow-up Observation Program Working Group (TFOPWG).  If you are a member of that group and wish to access that data, you must log on to ExoFOP.  
The php functions, which are URL calls, can be called as part of a curl or wget script if the use logs in and the login is saved as a cookie which is passed as part of the curl or wget call.  
Here is an example of how to do this with curl and wget.  

Create and save the cookie.  
wget:  
```
wget --keep-session-cookies --save-cookies ./mycookie.txt --post-data "username=username&password=password&ref=login_user&ref_page=/tess/" "https://exofop.ipac.caltech.edu/tess/password_check.php"
```  
curl:  
```
curl -b ./mycookie.txt -c ./mycookie.txt --data "username=myusername&password=mypassword&ref=login_user&ref_page=/tess/" https://exofop.ipac.caltech.edu/tess/password_check.php
```
  
  
Load the cookie to pass the login information.  
wget:  
```
wget --load-cookies ./mycookie.txt  [php call]
```
curl:  
```
curl -b ./mycookie.txt [php call]

```

### TOI

__

download_toi.php  
Accesses the full TOI table from https://exofop.ipac.caltech.edu/tess/view_toi.php  

Optional terms:  
toi=[nnn]  
Accesses a single TOI from the TOI table  

sort=[column]  
Available columns to sort by: 
tid, toidesc, toi, toiasc, master, sg1a, sg1b, sg2, sg3, sg4, sg5, acwg, tmag, planetrad, planetname, planetnum, planetinsol, planetteq, planetsnr, disposition, tfopwgdisp, source, ra, dec, pmra, pmdec, epoch, period, duration, depth, distance, teff, logg, rad, metallicity, sectors, toi_alerted, toi_edited, lastmod (default search)  

output=[text|csv|pipe]  
Sets output format, default is 'text'

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_toi?toi=119&output=pipe

### Target Overview

__

download_target.php?id=[nnnnnnnnn]  
Downloads composite text file of ExoFOP page  

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_target.php?id=278683844

### Files  

download_files.php?id=[nnnnnnnnn]  
Downloads all files for a given TIC id as a single tar  

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_files.php?id=278683844  

__

download_files_zip.php?id=[nnnnnnnnn]  
Downloads all files for a given TIC id as a single zip file  

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_files_zip.php?id=278683844  

__

download_filelist.php  
Downloads list of files associated with a given star containing following columns:  
Type, File Name, File ID, TIC, TOI, Date, User, Group, Tag, Description  

Required term (one of the following):  
id=[nnnnnnnnn] TIC ID  
toi=[nnn] TOI  


Optional terms:  
output=[csv]  
Default output is pipe delimited, downloaded file has .pipe extension  

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_filelist.php?id=278683844&output=csv  

__

download_tag_files.php?tag=[tag]  
Downloads all files for a given tag as a single tar  
Tag can be string or associated number  

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_tag_files.php?tag=1  

__

download_tag_files_zip.php?tag=[tag]  
Downloads all files for a given tag as a single zip  
Tag can be string or associated number  

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_tag_files_zip.php?tag=1  


### Nearby Target  

download_nearbytarget.php?id=[nnnnnnnnn]  
Accesses nearby targets for a given TIC ID and returns following columns:  
TIC ID, TOI, RA, Dec, TESS Mag, GAIA DR2, GAIA Mag, Distance (pc), Distance Err (pc), PM RA (mas/yr), PM RA Err (mas/yr), PM Dec (mas/yr), PM Dec Err (mas/yr), Separation (arcsec), Position Angle (deg E. of N.)  

Optional terms:  
output=[text|csv|pipe]  
text is default option  

sort=[column]  
Available columns to sort by: id (TIC ID), toi, ra, dec, pmra, pmdec, tmag, distance, separation (default search)   

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_nearbytarget.php?id=278683844&output=csv&sort=tmag

### Imaging Observation Summaries  

Imaging observation summaries are viewable at https://exofop.ipac.caltech.edu/tess/view_imaging.php and contain the following columns:  
TIC ID, TIC, TOI, Telescope, Instrument, Filter, Image Type, Pixel Scale, PSF, Contrast, Obs Date, User, Group, Tag, Notes  

__

download_imaging.php  
Accesses all imaging observation summaries and returns above columns

Optional terms:  
output=[text|csv|pipe]  
text is default option  

sort=[column]  
Available columns to sort by: id (TIC ID, default sort), tel, inst, filt, pix, imagetype, psf, cont, date, tag, notes, user, group  
  
Sample call:  
https://exofop.ipac.caltech.edu/tess/download_imaging.php?output=pipe&sort=user  

__

download_tag_imaging.php?tag=[tag]  
Accesses all imaging observation summaries for a given tag and returns above columns  
Tag can be string or associated number  

Optional terms:  
output=[text|csv|pipe]  
text is default option  

sort=[column]  
Available columns to sort by: id (TIC ID, default sort), tel, inst, filt, pix, imagetype, psf, cont, date, tag, notes, user, group  

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_tag_imaging.php?output=pipe&tag=95132  

### Spectroscopic Observation Summaries  

Spectroscopic observation summaries are viewable at https://exofop.ipac.caltech.edu/tess/view_spect.php and contain the following columns:  
TIC ID, TIC, TOI, Telescope, Instrument, Spectral resolution, Wavelength coverage, SNR/resolution, SNR wavelength, Appropriate to PRV, Obs Date, User, Group, Tag, Notes   
–

download_spect.php  
Accesses all spectroscopic observation summaries and returns above columns  

Optional terms:  
output=[text|csv|pipe]  
text is default option  

sort=[column]  
Available columns to sort by: id (TIC ID, default sort), tel, inst, res, cov, snrres, snrwave, prv, date, tag, notes, user, group  

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_spect.php?output=pipe&sort=user  
__

download_tag_spect.php?tag=[tag]  
Accesses all spectroscopic observation summaries for a given tag and returns above columns  
Tag can be string or associated number  

Optional terms:  
output=[text|csv|pipe]  
text is default option  

sort=[column]  
Available columns to sort by: id (TIC ID, default sort), tel, inst, res, cov, snrres, snrwave, prv, date, tag, notes, user, group  

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_tag_spect.php?output=pipe&tag=1494


### Time Series Observation Summaries

Time series observation summaries are viewable at https://exofop.ipac.caltech.edu/tess/view_tseries.php and contain the following columns:  
TIC ID, TIC, TOI, Telescope, Camera, Filter, Pix Scale (arcsec), PSF (arcsec), Phot Aperture Rad (pix), Date (UT), Duration (m), # of Obs, Type, Transit Coverage, Delta Mag, User, Group, Tag, Notes  

–

download_tseries.php  
Accesses all time series observation summaries and returns above columns  

Optional terms:  
output=[text|csv|pipe]  
text is default option  

sort=[column]  
Available columns to sort by: id (TIC ID, default sort), tel, camera, filt, pix, psf, photaprad, date, transcov, deltamag, tag, notes, user, group  

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_tseries.php?output=pipe&sort=user  

–

download_tag_tseries.php?tag=[tag]  
Accesses all time series observation summaries for a given tag and returns above columns  
Tag can be string or associated number  

Optional terms:  
output=[text|csv|pipe]  
text is default option  

sort=[column]  
Available columns to sort by: id (TIC ID, default sort), tel, camera, filt, pix, psf, photaprad, date, transcov, deltamag, tag, notes, user, group  

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_tag_tseries.php?output=pipe&tag=1

### Observing Notes  

download_obsnotes.php  

Returns the following columns:  
ID, TIC ID, Username, Groupname, TAG ID, Lastmod, notes  

Required term (one of the following):  
tag=[tag]  
Tag can be string or associated number  

tid=[nnnnnnnnn]  
tid is TIC ID  

id=[nnnn]  
id is row ID associated with note  


Optional terms:  
output=[pipe]  
Default is csv
  
Sample call:  
https://exofop.ipac.caltech.edu/tess/download_obsnotes.php?tag=3059

### Tag Overview

__

download_tag.php?tag=[tag]  
Downloads text file containing tag string and tag ID  
Also returns any of the following tables that have associated data:  
TFOPWG Disposition, Planet Parameters, Stellar Paramters, Magnitudes, Imaging Observations, Spectroscopic Observations, Time Series Observations, Files  

Observing Notes are not returned  
  
Sample call:  
https://exofop.ipac.caltech.edu/tess/download_tag.php?tag=1  

__

download_user_tags.php?username=[username]  
If username parameter not included, search is conducted on logged-in user's username  
Accesses list of user tags and returns following columns:  
User, Tag, Tag ID, Number of Uploaded Time Series Observations, Number of Uploaded Spectroscopy Observations, Number of Uploaded Imaging Observations, Number of Uploaded Files, Number of Uploaded Observaton Notes, Number of Uploaded Stellar Parameters, Number of Uploaded Stellar Companions, Number of Uploaded Planet Parameters, Date Tag Created, Date Tag Last Modified  

Optional terms:  
output=[text|csv|pipe]  
text is default option  

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_user_tags.php?username=ciardi  

### Stellar Companions  

Stellar companions are viewable at https://exofop.ipac.caltech.edu/tess/view_stellarcomp.php and includes all companions reported, not necessarily only those stellar companions that are gravitationally bound. This contains the following columns:  
TIC ID, TOI, Separation (arcsec), Separation (arcsec) Error, Pos Angle E of N (deg), Pos Angle E of N (deg) Error, Filter, Delta Mag, Delta Mag Error, Observation Date (UT), Date, User, Group, Tag, Notes

_

download_stellarcomp.php  
Accesses all user-reported stellar companions and returns the above columns  

Optional terms:  
output=[text|csv|pipe]  
text is default option  

sort=[column]
Available columns to sort by: id (TIC ID, default sort), lastmod (descending)  

Sample call:  
https://exofop.ipac.caltech.edu/tess/download_stellarcomp.php?output=pipe
